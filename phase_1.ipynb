    <div class="paragraph">
        <br>Data Cleaning process broadly involves 2 sub-processes, handling <strong>Null values</strong> and handling <strong>Outliers</strong>.
    </div>
    <div class="subsubsection-title", style="font-size: 18px;"><br><br>2.1.1 Null Values</div>
    <div class="paragraph">
        <br> Null values, or missing data, can skew analysis and lead to incorrect conclusions. Instead of ignoring null values, we explore various imputation techniques to replace them with suitable values:
    </div>
    <ul>
        <li class="list-item"><strong>Mean Imputation:</strong> Replacing null values with the mean of the non-null values. This is useful for continuous data without outliers.</li><br>
        <li class="list-item"><strong>Median Imputation:</strong> Replacing null values with the median of the non-null values. This is effective for skewed distributions or when outliers are present.</li><br>
        <li class="list-item"><strong>Mode Imputation:</strong> Replacing null values with the mode (most frequent value) of the non-null values. This is suitable for categorical data.</li>
    </ul>
    <div class="paragraph">
        <strong>Exploring Correlations:</strong><br>
        While handling null values, it is essential to explore correlations between attributes to maintain the integrity of the data. For instance, if a null value in one attribute correlates strongly with another attribute, using the latter can guide more accurate imputation. Correlation matrices and heatmaps can help identify these relationships.
    </div>

    <div class="subsubsection-title", style="font-size: 18px;">2.1.2 Outliers</div>
    <div class="paragraph">
        <br>Outliers can distort statistical analyses and provide misleading insights. Various methods to detect and handle outliers include:
    </div>
    <ul>
        <li class="list-item"><strong>Standard Deviation Method:</strong> Identifying outliers as values beyond a certain number of standard deviations from the mean. Typically, values more than 3 standard deviations away are considered outliers.<br></li><br>
        <li class="list-item"><strong>Interquartile Range (IQR) Method:</strong> Calculating the IQR (difference between the 75th and 25th percentiles) and identifying outliers as values below Q1 - 1.5IQR or above Q3 + 1.5IQR.<br></li><br>
        <li class="list-item"><strong>Z-score Method:</strong> Calculating the Z-score (number of standard deviations a data point is from the mean). Values with a Z-score greater than 3 or less than -3 are considered outliers.<br></li><br>
        <li class="list-item"><strong>Percentile Method:</strong> Identifying outliers based on percentile thresholds, e.g., below the 1st percentile or above the 99th percentile.</li>
    </ul>
    <br>
    <div class="paragraph">
        <strong>Statistical and Domain Knowledge:</strong><br>
        While theoretical methods are essential, it's crucial to consult with business managers to understand the context of the data. Combining statistical methods with domain expertise ensures data cleaning is practical and relevant to business objectives.
    </div>

<h3 id="f:eqn", style="font-size: 20px;">2.3 Statistical Concepts for A/B Testing</h3>

<ul>
    <li><strong>Random Sampling:</strong> Selecting a subset of individuals from a population to represent the entire population.</li>
    <li><strong>Sampling Bias:</strong> Occurs when the sample is not representative of the population, leading to skewed results.</li>
    <li><strong>Law of Large Numbers:</strong> As the sample size increases, the sample mean will get closer to the population mean.</li>
    <li><strong>Central Limit Theorem:</strong> The distribution of the sample mean approaches a normal distribution as the sample size grows, regardless of the population's distribution.</li>
    <li><strong>Standard Error:</strong> The standard deviation of the sample mean distribution, indicating the accuracy of the sample mean as an estimate of the population mean.</li>
    <li><strong>Z-score Table:</strong> A table that provides the probability of a standard normal variable being below a particular value, used to calculate p-values.</li>
    <li><strong>Confidence Interval:</strong> A range of values derived from the sample mean that is likely to contain the population mean.</li>
    <li><strong>Hypothesis Testing:</strong> A statistical method to test an assumption regarding a population parameter.</li>
    <li><strong>Z-test:</strong> Used to determine if there is a significant difference between the means of two groups, assuming a normal distribution and known variance.</li>
    <li><strong>Significance Level (&alpha;):</strong> The probability threshold below which the null hypothesis is rejected, typically set at 0.05.</li>
    <li><strong>Type 1 Error (False Positive):</strong> Rejecting the null hypothesis when it is true.</li>
    <li><strong>Type 2 Error (False Negative):</strong> Failing to reject the null hypothesis when it is false.</li>
    <li><strong>P-value:</strong> The probability of observing test results at least as extreme as the results actually observed, under the assumption that the null hypothesis is true.</li>
    <li><strong>Effect Size:</strong> A measure of the strength of the relationship between two variables or the extent of an intervention.</li>
    <li><strong>A/B Testing:</strong> Comparing two versions (A and B) to determine which one performs better statistically. This involves running experiments on different groups and analyzing the outcomes using the above statistical concepts to make informed decisions.</li>
</ul>

We can handle these null values using different ways,

1. **Remove them**: Since there are 50 of them in a dataframe of 1000, we will not remove them as we don't want to loose some important records
1. **Replace them with mean or median**: It is suggested with use median in the case of income. This is because in an income data there could be outliers and median is more robust to these outliers
1. **Replace them with median per occupation**: Occupation wise median income can vary. It is best to use a median per occupation for replacement

Before replacing the null values with the median annual_income value we shall see if there is any correlation of annual_income with other attributes

Since we cannot get correlation coefficient between numerical and non-numerical attributes we will rather see the variation of annual income with other attributes by data visualisation techniques.

```python
# Setting up subplots with 2 rows and 2 columns
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Define a palette for colors
palette = 'Set2'

# Plotting median annual_income by gender
sns.barplot(x='gender', y='annual_income', data=df_cust, ax=axes[0, 0], palette=palette, estimator=np.median)
axes[0, 0].set_title('Median Annual Income by Gender')

# Plotting median annual_income by location
sns.barplot(x='location', y='annual_income', data=df_cust, ax=axes[0, 1], palette=palette, estimator=np.median)
axes[0, 1].set_title('Median Annual Income by Location')

# Plotting median annual_income by occupation with hue='occupation'
sns.barplot(x='occupation', y='annual_income', data=df_cust, ax=axes[1, 0], palette=palette, hue='occupation',
            estimator=np.median, order=df_cust.groupby('occupation')['annual_income'].median().sort_values().index)
axes[1, 0].set_title('Median Annual Income by Occupation')
axes[1, 0].tick_params(axis='x', rotation=45)

# Plotting median annual_income by marital_status with hue='marital_status'
sns.barplot(x='marital_status', y='annual_income', data=df_cust, ax=axes[1, 1], palette=palette, hue='marital_status',
            estimator=np.median, order=['Single', 'Married'])
axes[1, 1].set_title('Median Annual Income by Marital Status')

# Adjusting layout
plt.tight_layout()

# Display the plot
plt.show()
```

We can observe that the variation of median 'annual_income' is more with 'occupation' compared to other non-numerical attributes

We will replace these null values with corresponding values of 'median by occupation' , since median value is such a measure of central tendency that it is unaffected by outliers unlike mean.


```python
# Calculate median income by occupation
median_income_by_occupation = df_cust.groupby('occupation')['annual_income'].transform('median')

# Replace null values in annual_income with median income by occupation
df_cust['annual_income'].fillna(median_income_by_occupation, inplace=True)
```

Here we can see clear relationship between credit score and credit limit. Where there are levels for example, upto 650 score is getting a very minor credit limit (<1000$) where as a score between 650 to 700 is getting around 20000. Score between 700 to 750 is getting around 40K etc.

```python
bins = [300, 450, 500, 550, 600, 650, 700, 750, 800]
labels = ['300-449', '450-499', '500-549', '550-599', '600-649', '650-699', '700-749', '750-800']

# Add 'credit_score_range' column based on credit score ranges
df_cs['credit_score_range'] = pd.cut(df_cs['credit_score'], bins=bins, labels=labels, right=False)
df_cs.head()
```

We can now see a new column called credit_score_range which is calculated based on the credit_score column

```python
mode_df = df_cs.groupby('credit_score_range')['credit_limit'].apply(lambda x: x.mode()[0]).reset_index(name='mode_credit_limit')
mode_df
```

We will replace null values in 'credit_limit' with mode_credit_limit corresponding to its credit_score_range

```python
merged_df = pd.merge(df_cs, mode_df, on='credit_score_range', how='left')

# Replace null values in 'credit_limit' with mode_credit_limit
merged_df['credit_limit'] = merged_df.apply(lambda x: x['mode_credit_limit'] if pd.isnull(x['credit_limit']) else x['credit_limit'], axis=1)

# Drop the 'mode_credit_limit' column as it's no longer needed
merged_df.drop(columns='mode_credit_limit', inplace=True)

df_cs=merged_df
```
